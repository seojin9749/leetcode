# 1. LeetCode 225. Implement Stack using Queues
push(x)는 간단하게 q1에 추가합니다.
pop()은 q1에서 마지막 하나만 남기고 q2로 옮깁니다. 마지막 하나는 스택의 top이므로 제거합니다.
top()은 pop()과 거의 같지만, 마지막 원소를 다시 q2에 넣어줍니다.
항상 마지막에 q1, q2를 swap해서 q1이 다시 메인 큐가 되게 합니다.

```python
from collections import deque

class MyStack:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q1.append(x)

    def pop(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        val = self.q1.popleft()
        self.q1, self.q2 = self.q2, self.q1
        return val

    def top(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        val = self.q1.popleft()
        self.q2.append(val)
        self.q1, self.q2 = self.q2, self.q1
        return val

    def empty(self) -> bool:
        return not self.q1
```

# 2. LeetCode 232. Implement Queue using Stacks

## 🔸 문제 설명
두 개의 스택만을 이용해 Queue (FIFO)를 구현해야 합니다.  
다음 연산을 구현해야 합니다:
- `push(x)` : 원소 x를 큐에 삽입
- `pop()` : 큐의 맨 앞 원소를 제거
- `peek()` : 큐의 맨 앞 원소를 확인
- `empty()` : 큐가 비었는지 확인

## 🔹 접근 방법
- 두 개의 스택(`in_stack`, `out_stack`)을 사용
- `push`: `in_stack`에 원소 삽입
- `pop` / `peek`: `out_stack`에 원소가 없으면 `in_stack`의 원소를 모두 꺼내 반대로 넣어준다
  → 이렇게 하면 FIFO 순서가 유지됨

## ⏱️ 시간 복잡도
- `push`: O(1)
- `pop`, `peek`: 평균적으로 O(1) (Amortized)
- `empty`: O(1)


```python
class MyQueue:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        self.move()
        return self.out_stack.pop()

    def peek(self) -> int:
        self.move()
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack

    def move(self):
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
```

```python
#3-1 문제
class ListQueue:
    def __init__(self):
        self.__queue = []

    def enqueue(self, x):
        # 앞쪽(index 0)에 요소를 삽입 → tail
        self.__queue.insert(0, x)

    def dequeue(self):
        # 뒤쪽에서 꺼냄 → front
        return self.__queue.pop()

    def front(self):
        # 마지막 요소가 front
        return self.__queue[-1]

    def is_Empty(self) -> bool:
        return len(self.__queue) == 0

    def dequeue_All(self):
        self.__queue.clear()
```

```python
#3-2
from collections import deque

def is_in_set(s: str) -> bool:
    queue = deque(s)  # 문자열을 큐에 문자 단위로 저장

    if not queue:
        return False

    # 첫 글자가 $인지 확인
    if queue.popleft() != '$':
        return False

    # 이후에는 어떤 문자열(w)이든 상관없으므로 True
    return True


# 테스트
test_cases = ['$', '$abc', '$123', 'abc$', 'abc', '']
for s in test_cases:
    print(f"{s!r} => {is_in_set(s)}")
```

```python
#3-3
class Node:
    def __init__(self, item, link=None):
        self.item = item
        self.link = link

class LinkedQueue:
    def __init__(self):
        self.front = self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = self.rear = new_node
        else:
            self.rear.link = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            return None
        item = self.front.item
        self.front = self.front.link
        if self.front is None:
            self.rear = None
        return item

    def peek_all(self):
        current = self.front
        items = []
        while current:
            items.append(current.item)
            current = current.link
        return items

def copy_queue(a: LinkedQueue) -> LinkedQueue:
    b = LinkedQueue()
    temp = []
    while not a.is_empty():
        item = a.dequeue()
        temp.append(item)
        b.enqueue(item)

    for item in temp:
        a.enqueue(item)

    return b
```

```python
#3-4
from collections import deque

class StackWithTwoQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        return self.q1.popleft() if self.q1 else None

    def top(self):
        return self.q1[0] if self.q1 else None

    def empty(self):
        return not self.q1
```

```python
#3-5
class QueueWithTwoStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        while self.stack1:
            self.stack2.append(self.stack1.pop())
        self.stack1.append(x)
        while self.stack2:
            self.stack1.append(self.stack2.pop())

    def dequeue(self):
        if not self.stack1:
            return None
        return self.stack1.pop()

    def front(self):
        return self.stack1[-1] if self.stack1 else None
```

```python
#3-8
class ListDeque:
    def __init__(self):
        self.__queue = []

    def is_empty(self) -> bool:
        return len(self.__queue) == 0

    def enqueue_front(self, x):
        self.__queue.insert(0, x)  # 앞에 삽입 (비효율적 - O(n))

    def enqueue_rear(self, x):
        self.__queue.append(x)     # 뒤에 삽입 (효율적 - O(1))

    def dequeue_front(self):
        if self.is_empty():
            return None
        return self.__queue.pop(0)  # 앞에서 삭제 (O(n))

    def dequeue_rear(self):
        if self.is_empty():
            return None
        return self.__queue.pop()   # 뒤에서 삭제 (O(1))

    def front(self):
        if self.is_empty():
            return None
        return self.__queue[0]

    def rear(self):
        if self.is_empty():
            return None
        return self.__queue[-1]

    def deque_all(self):
        self.__queue.clear()

    def print_deque(self):
        print("Deque from front:", end=' ')
        for item in self.__queue:
            print(item, end=' ')
        print()
```
