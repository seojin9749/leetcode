```python
from collections import deque

class MyStack:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x: int) -> None:
        self.q1.append(x)

    def pop(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        val = self.q1.popleft()
        self.q1, self.q2 = self.q2, self.q1
        return val

    def top(self) -> int:
        while len(self.q1) > 1:
            self.q2.append(self.q1.popleft())
        val = self.q1.popleft()
        self.q2.append(val)
        self.q1, self.q2 = self.q2, self.q1
        return val

    def empty(self) -> bool:
        return not self.q1
```

## ğŸ”¹ ì ‘ê·¼ ë°©ë²•
- ë‘ ê°œì˜ ìŠ¤íƒ(`in_stack`, `out_stack`)ì„ ì‚¬ìš©
- `push`: `in_stack`ì— ì›ì†Œ ì‚½ì…
- `pop` / `peek`: `out_stack`ì— ì›ì†Œê°€ ì—†ìœ¼ë©´ `in_stack`ì˜ ì›ì†Œë¥¼ ëª¨ë‘ êº¼ë‚´ ë°˜ëŒ€ë¡œ ë„£ì–´ì¤€ë‹¤

## â±ï¸ ì‹œê°„ ë³µì¡ë„
- `push`: O(1)
- `pop`, `peek`: í‰ê· ì ìœ¼ë¡œ O(1) (Amortized)
- `empty`: O(1)


```python
class MyQueue:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x: int) -> None:
        self.in_stack.append(x)

    def pop(self) -> int:
        self.move()
        return self.out_stack.pop()

    def peek(self) -> int:
        self.move()
        return self.out_stack[-1]

    def empty(self) -> bool:
        return not self.in_stack and not self.out_stack

    def move(self):
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
```

```python
#3-2
from collections import deque

def is_in_set(s: str) -> bool:
    queue = deque(s)  # ë¬¸ìì—´ì„ íì— ë¬¸ì ë‹¨ìœ„ë¡œ ì €ì¥

    if not queue:
        return False

    # ì²« ê¸€ìê°€ $ì¸ì§€ í™•ì¸
    if queue.popleft() != '$':
        return False

    # ì´í›„ì—ëŠ” ì–´ë–¤ ë¬¸ìì—´(w)ì´ë“  ìƒê´€ì—†ìœ¼ë¯€ë¡œ True
    return True


# í…ŒìŠ¤íŠ¸
test_cases = ['$', '$abc', '$123', 'abc$', 'abc', '']
for s in test_cases:
    print(f"{s!r} => {is_in_set(s)}")
```

```python
#3-3
class Node:
    def __init__(self, item, link=None):
        self.item = item
        self.link = link

class LinkedQueue:
    def __init__(self):
        self.front = self.rear = None

    def is_empty(self):
        return self.front is None

    def enqueue(self, item):
        new_node = Node(item)
        if self.is_empty():
            self.front = self.rear = new_node
        else:
            self.rear.link = new_node
            self.rear = new_node

    def dequeue(self):
        if self.is_empty():
            return None
        item = self.front.item
        self.front = self.front.link
        if self.front is None:
            self.rear = None
        return item

    def peek_all(self):
        current = self.front
        items = []
        while current:
            items.append(current.item)
            current = current.link
        return items

def copy_queue(a: LinkedQueue) -> LinkedQueue:
    b = LinkedQueue()
    temp = []
    while not a.is_empty():
        item = a.dequeue()
        temp.append(item)
        b.enqueue(item)

    for item in temp:
        a.enqueue(item)

    return b
```

```python
#3-4
from collections import deque

class StackWithTwoQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q2.append(x)
        while self.q1:
            self.q2.append(self.q1.popleft())
        self.q1, self.q2 = self.q2, self.q1

    def pop(self):
        return self.q1.popleft() if self.q1 else None

    def top(self):
        return self.q1[0] if self.q1 else None

    def empty(self):
        return not self.q1
```

```python
#3-5
class QueueWithTwoStacks:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, x):
        while self.stack1:
            self.stack2.append(self.stack1.pop())
        self.stack1.append(x)
        while self.stack2:
            self.stack1.append(self.stack2.pop())

    def dequeue(self):
        if not self.stack1:
            return None
        return self.stack1.pop()

    def front(self):
        return self.stack1[-1] if self.stack1 else None
```
